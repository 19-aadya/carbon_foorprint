```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/supply_chain_emission_factors.csv")

# Drop nulls and look at the head
df = df.dropna()
df.head()
print(df.columns)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select top 20 industries by emission factor
top_n_df = df.nlargest(20, "Supply Chain Emission Factors without Margins")

# Use a horizontal bar plot
plt.figure(figsize=(12, 8))
sns.barplot(data=top_n_df, y="2017 NAICS Title", x="Supply Chain Emission Factors without Margins")

plt.xlabel("Emission Factor (kg CO2e/2022 USD)")
plt.ylabel("Industry")
plt.title("Top 20 Industries by Emission Factor")
plt.tight_layout()
plt.show()
```
